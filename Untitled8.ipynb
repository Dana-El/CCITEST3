```python
def calculate_discounted_prices(prices, discount_percentage):
    """
    Calculates discounted prices for a list of items.

    Args:
        prices (list): A list of original prices.
        discount_percentage (float): The discount rate (e.g., 10 for 10%).

    Returns:
        list: A list of discounted prices.
    """
    discount_factor = 1 - (discount_percentage / 100)
    discounted_prices = [price * discount_factor for price in prices]
    return discounted_prices
```

This `calculate_discounted_prices` function takes a list of original prices and a discount percentage. It returns a new list with the prices after applying the discount.

```python
# Example usage:
original_prices = [100, 115, 114, 118.7, 50, 200]
discount = 15 # 15% discount

discounted_items = calculate_discounted_prices(original_prices, discount)

print(f"Original prices: {original_prices}")
print(f"Discount percentage: {discount}%")
print(f"Discounted prices: {discounted_items}")
```
